### Featires of mongodb

1. Adhoc querries: In mongo db we dont have an idea how our table structures will be like sql we dump what ever data is coming. so we dont have a idea of querry structure in future.
2. Aggrigation : similar to sql mongo db also can work in grouped data
3. schema less db: table dont have a particular structure, what ever data is coming we dump it so no schema. 
4. document oriented: instead of row in sql here its document, each document has a unique identifier.
5. indexing: for fetching data faster.
6. high performance: provide high avaliablity, scaliablity and better query response.
    - high avaliablity: in mongo multiple copies of data in multiple places is created automatically, so fetching is so easy and data easily avaliable. so one issue happen in one place we can access from another place.
    - scaliablity: MongoDB has the ability to store both sharded and unsharded collections in a sharded cluster. This allows the application to take full advantage of the cluster for large data sets while using a primary shard for small data sets.
