### **Identify and Handling Missing Values**

**<span style="color:green">How to Deal With Missing Values?</span>**

- **Check with the data collection sources we may get the hint from there.**

- **Drop the missing values, and we have 2 options.**

    - **Drop the entire column/feature: if a column have most of the values are null the we can drop that column.**

    - **Drop the entire row: if we have lots of data and only some of datapoints have null values its ok to remove that datapoints.**

- **Replace the missing values.**

    - **Replave with AVG value.**

    - **Replace with Mode.**

    - **Replace with custom logic.**

- **Leave Null as it is, some situations we have to keep null as it is.**

**<span style="color:green">How to Drop Missing Values in Python ?</span>**

- **DataFrame.dropna(subset=["Price"],axis=0,inplace=True)  , Here drops entire row where Price is null.**

- **axis=0 drops entire rows.**

- **axis=1 drops entire column**

**<span style="color:green">How to Replace Missing Values in Python ?</span>**

- **DataFrame['Price'].replace(np.nan,DataFrame['Price'].mean())**

- **DataFrame['Make'].replace(np.nan,DataFrame['Make'].mode())**

### **Data Normalization**

**<span style="color:red">Why Normalizing is required?</span>**

**<span style="color:green">Methods of Normalizing Data</span>**

**Simple Feature Scaling.**


\begin{align}
\large
X_{new} = \frac{X_{old}}{X_{max}}
\end{align}

**Min-Max Scaling**


\begin{align}
\large
X_{new} = \frac{X_{old} - X_{min}}{X_{max}-X_{min}}
\end{align}

**Z-Score**


\begin{align}
\large
X_{new} = \frac{X_{old} - \mu}{\sigma}
\end{align}
