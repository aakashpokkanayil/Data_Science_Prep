```python
test_scr_dist = %sql select test_score as "Test Score",count(*) as "Frequency" from INTERNATIONAL_STUDENT_TEST_SCORES group by test_score;
test_scr_dist
```

```python
df_scr_dist=test_scr_dist.DataFrame()
df_scr_dist.head()
```
