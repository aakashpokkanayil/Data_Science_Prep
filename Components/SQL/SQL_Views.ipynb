# **What is a Views ?**

- #### **View is one of the way to show data from one or more tables, it is a virtual table based on a resultset of an sql statement.**

- #### **View can have specified columns from single or multiple tables even from other views.**

- #### **Once Created, can be queried like a table.**

- #### **DML like insert, update, delete on view(single base tabled view) will affect base table also.**

- #### **When we define a view its define tio is only stored not the data , data will be directly linked to base table data.**

- #### **<span style="color:red">View can't create with query contains order by clause</span>**

- #### **We can remove view by DROP command**

# **When should you use a view?**

- #### **when we need to display only non-sensitive data from tables**

- #### **We can combine one or more tables and represent as a single table, user will only felt like accessing a table.**

- #### **Simplify access to data.**
