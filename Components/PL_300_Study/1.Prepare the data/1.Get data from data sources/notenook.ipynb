**How to connect to a data source?**

Home(Menu) --> Get Data (option) --> Select required datasource from poped up list --> Browse the file from its location --> So preview of data will be loaded --> we can either select load option  or select transfrom option.

**How to change data source of a loaded data?**

1. If we directly change the directory via file explorer , we will get error from powerBI after refreshing, that is  **file could not found**  
2. So we have to update source settings in power BI as well.
3. From Home menu --> In Transfrom Data option(select dropdown menu button) --> Select Data source settings -->from the poped window select required data to change path --> select change source settings -->  then again a window will poped up there we can change file path --> after that in Home --> Refresh
