```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
df1=pd.read_csv('iris.csv')

counts,bin_edges=np.histogram(df1.PetalLengthCm,bins=10,density=True)
```

```python
pdf=counts/sum(counts)
print(pdf)
```

```python
cdf=np.cumsum(pdf)
print(cdf)
```

```python
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()
```
