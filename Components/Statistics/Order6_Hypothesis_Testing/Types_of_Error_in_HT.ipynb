# Types of Error

Type I : Ha is False but Ha Accepted(H0 rejected).

Type II : Ha is true but Ha Rejected(Failed to reject H0).

always think in terms of Ha , for easy ness.




### Type 1 error = Significance Level

1. In a hypothesis test, a type I error occurs when you reject a null hypothesis that is actually true. In other words, **a statistically significant test(Hypo test) result suggests that a population effect exists but, in reality, it does not exist.** The difference you observed in the sample is the product of random sample error(error caused by samples).

2. The probability of committing a type I error equals the **significance level you set for your hypothesis test.** 

3. A significance level of 0.05 indicates that you are willing to accept a 5% chance (For type I error) that you are wrong when you reject the null hypothesis. **To lower this risk, you must use a lower value for alpha.**

4. However, if you use a lower value for alpha, **you are less likely to detect a true difference if one really exists.**

## Type II error = Beta

1. Type II error occur when you failed to reject H0 when that actually is False(Ha True)

2. Type two error causes due to Increase in Varience and sample size in data.

3. Prob of rejecting a False H0(FN) = 1-beta, 1-beta aka **Power of the test**

4. To Increase Power of test(1-beta) or to reduce Type II error we have to increase sample size (varience we cant control that much).

5. Type I and Type II error are inversily proportional.
