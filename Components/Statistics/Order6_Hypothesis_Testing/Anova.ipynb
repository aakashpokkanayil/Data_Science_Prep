# Anova Test

1. Anova : Analysis of varience

2. its an extension of z-test/t-test

3. In anova test Varience are analysed.

4. Types of anova:

    1. 1 way anova: comparison based on One factor
    
    2. 2 way anova: comparison based on Two factor
    
    3. n way anova: comparison based on n factor
      
5. F-statistics

    1. its is used to compair varience between two groups
    
    2. F= Varience of first group/ varience of 2nd group.  first group is select with highiest varience.
