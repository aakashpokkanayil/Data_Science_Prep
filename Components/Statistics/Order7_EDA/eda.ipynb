# Visualization

```python
ax=sns.countplot(x=Ipldata.AGE)
for label in ax.containers:
    ax.bar_label(label)
plt.show()
```

```python
sns.histplot(x=Ipldata['SOLD PRICE'])
plt.show()
```

```python
sns.histplot(x=Ipldata['SOLD PRICE'],bins=20)
plt.show()
```

```python
sns.displot(x=Ipldata['SOLD PRICE'],kind='kde')
plt.show()
```

```python
sns.distplot(x=Ipldata['SOLD PRICE'])
plt.show()
```

```python
sns.boxplot(x=Ipldata['SOLD PRICE'])
plt.show()
```

```python
box=plt.boxplot(Ipldata['SOLD PRICE'])
plt.show()
```

```python
[ i.get_ydata()[0] for i in box['caps']]  # UL AND LL in accordance with iqr outlier detection formula
```

```python
[i.get_ydata()[0] for i in box['whiskers']] # q1 and q3
```

```python
[item.get_ydata()[0] for item in box['medians']]# q2 or median
```

```python
sns.barplot(x='AGE',y='SOLD PRICE', data=Ipldata)
plt.show()
```

```python
sns.barplot(x='AGE',y='SOLD PRICE', data=Ipldata,ci=None)
plt.show()
```

```python
sns.barplot(x='AGE',y='SOLD PRICE', data=Ipldata,ci=None,hue='PLAYING ROLE')
plt.show()
```
