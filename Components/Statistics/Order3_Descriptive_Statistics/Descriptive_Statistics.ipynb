### Measure of Central Tentency (Single Variate)

1. **measure of central tendency** gives us **Statistical CenterPoint** of a Feature in a dataset


2. These measures indicate where __most values in a distribution fall__ and are also referred to as the __central location of a distribution.__


3. If a data follows **normal distribution(bell curve)** then **around central tentency majority value will fall.** 


4. **Mean, Median and Mode are the measure** which we used to identify **Central Tentency.**


5. CT will effect by all Arithemetic operation.



#### Mean

   1. **Average value** of the set of Numbers.
   

   2. Mean is a **number which denotes center value** of a data.
   
 
   3. Denoted by **μ** for population mean, **x̄** and for sample mean.
   

   4. __Center by Value.__ So it will be effected by __outlier.__
   
   
   5. It changes when **symmetricity of the distribution/graph is affected.**
   
   
#### Median

   1. Median is the value which __divides the data in 2 equal parts__ i.e. number of terms on right side of it is same as number of terms on left side of it when data is __arranged in either ascending or descending order.__
   
   
   2. Median is a **number which denotes center position** of a data.
   

   3. **Note:** If you sort data in descending order, it won’t affect median but __IQR will be negative.__ So ordering data in ascending order will be better.
   

   4. __Center by term/Position__. So __Outliers__ least effect Median.
   

   5. To find the median, Sort the data in ASC/DESC then if the N(number of observations) odd then median= (N+1)/2 Th item if N is even then Median= Avg((N/2) th item,(N/2)+1 th item).
   
   
#### Mode


   1. Mode is the term appearing maximum time in data set i.e. term that has **highest frequency.**
   
   
   2. Mode is Mostly **used on Categorical features.** Mode on Quantitative values make no sence.
   
   
   3. A dataset can be unimodal or multimodal.

### Measures of Dispersion or Spread or Variability (Single Variate)

1. STD and Varience gives **spread according to the mean.**


2. IQR and Range gives **spread from start to end of data.**


3. Measure of spread will only effect by Multiplication and division.


#### Standard Deviation

1. Standard deviation is the measurement of **average distance between each datapoint to the mean.**


2. it shows how data is **spread out around the mean.**  


3. A **low standard deviation** indicates that the data points tend to be close to the mean of the data set.


4. A **high standard deviation** indicates that the data points are spread out over a wider range of values.

#### IQR (Interquartile Range)

1. In statistics and probability, quartiles are values that divide your data into quarters provided data is sorted in an ascending order.


2. In a **Normal Distribution highest probability to have a value appeared heris in IQR.** it is the area where 50% of data resides, if i pick a point there is highest possiblity that that point comes in IQR.


3. IQR = Q3 – Q1.


4. Not effected by outliers also can be use for dectecting outliers.

# Degree of Freedom

1. when we study sample there is will always an error in sample mean to Pop mean, coz sample is only a subset of pop there will be always a diffrence in their mean.

2. as the sample size increase these error reduces

3. other thing we can do is increase the spread by decreasing the denominator of STD. by increasing spread of sample increases chances of sample include pop mean is increases.

4. how much we need to decrease demonimator of STD is calculated by Degree of freedom.

5. __We already know that the sum of the deviations about the mean, (xi - x'), must equal zero. Therefore, if the sample mean is known and first n - 1 observations are known, then the nth observation must be the value that causes the sum of the deviations to equal zero. We call n - 1 the degrees of freedom because the first n - 1 observations have freedom to be whatever value they wish, but the nth value has no freedom. It must be whatever value forces the sum of the deviations about the mean to equal zero.__

6. __n Other Words We have n - 1 degrees of freedom in the computation of S because an unknown parameter, Mue, is estimated with x . For each parameter estimated, we lose 1 degree of freedom.__

7. Degree of freedom is used when we are __estimating__ unknown characteristics of Population by using sample. tipically when we estimating not all observations in sample are free to vary due to mathematical restriction.

8. when n deviations about the sample mean are used to estimate the variablity in the population, only n-1  observations are free to vary. As a result there is only n-1 degree of freedom due to zero sum restriction(sum(xi - x')=0). when estimating standard deviation if we use n itself it would tends to underestimate vaiability of population. __More accurate result can be obtained if we use No of Independent obervation or deviations in the denominator that is degree of freedom. Important is That for each unknown parameter we will lose 1 degree of freedom. __

9. So as an colcusion we can say when we estimate pop from sample using dof will provice more accuracy.

### Measure of Symmetricity/Shape and Distribution

#### Normal Distribution AKA Gaussian distribution.


1. **Normal distribution**, also known as the **Gaussian distribution, is a probability distribution that is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean.**


2. In graphical form, the normal distribution appears as a **"bell curve".**


3. A normal distribution will have **zero skewness and a kurtosis of 3.**


4. A normal distribution with mean zero and the standard deviation is 1 is known as **Standard Normal Distribution.** 


5. In a Perfect Normal Distribution Mean (average), median (midpoint), and mode (most frequent observation) are all **equal to one another.** Also all three of these parameters will be points to **Center of Data** 


6. A Perfect Normal Distribution falls **symmetrically around the mean,** the **width of which is defined by the standard deviation.**


7. we can use skewness to identify Asymmetricity also kutosis indirectly points to symmetricity itself.

#### Skewness (Single Variate)


1. Skewness is usually described as a measure of a __dataset’s symmetry – or lack of symmetry.__


2. skewness calculate **how much a distribution is skewed compairing to standard Normal distribution.**


3. **Bell curve for Normal Distrinution** is the standard which we take for compairing skewness.


4. for a Bell curve LHS and RHS of mean will be symmetrical.


5. for a distribution if the __bell curve shifted to left side its called Right skewed distribution__


6. for a distribution if the __bell curve shifted to Right side its called Left skewed distribution__


7. if value of skewness is **+ve then its Right skewed**


8. if value of skewness is **-ve then its Left skewed**


9. If the skewness is __between -0.5 and 0.5, the data are fairly symmetrical.__


10. If the skewness is **between -1 and – 0.5 or between 0.5 and 1, the data are __moderately skewed.**


11. If the skewness is **less than -1 or greater than 1, the data are __highly skewed.__**

#### Kurtosis (Single Variate)

1. Kurtosis measures the __tail-heaviness or Peakness(old time peakness was focused but noew its about tail heavyness)__ of the distribution.


2. Normally **Kurtosis Value based on 3** but to covert the base in to 0 **excess kurtosis** Introduced.


3. As per kurtosis:

    a.  A distribution with kurtosis <3 (excess kurtosis <0) is called __platykurtic(Low kurtosis)__. 
    
    b. For __platykurtic, tails will be thinner(meaning that they have few outliers) and its central peak is lower and broader.__
        
    c. __platykurtic__ data contain mostly unique values, so it will be harder to find an pattern for ML algorithm.    
    
    d. **Thinner Tails means propability of extreme values are low , means less outliers**
        
    e. A distribution with kurtosis >3 (excess kurtosis >0) is called __leptokurtic(high Kurtosis)__.    
    
    f. Compared to a normal distribution, __leptokurtic tails are fatter(meaning that they have many outliers), and often its central peak is higher and sharper.__    
    
    g. **Fatter Tails means propability of extreme values are High , means more outliers**
       
    d. In __leptokurtic__ Repeated Data will be much higher that normal due to shrinked bell curve(probability of getting a value is much higher),these type of data is not good for ML Model, coz repeated data shows only the bias.
        
    e. if Kurtosis =3 or excess kurtosis = 0 then its __mesokurtic. which show its standard normal distribution.__ It keeps the balance between repetation of data and uniqueness of data.

```python
import numpy as np

x=[13,15,7,3,10,27]

y=[53,69,92,10,85,99]

x_mean=np.mean(x)

y_mean=np.mean(y)

sigma_x= np.sqrt(sum((x-x_mean)**2)/(len(x)-1))

sigma_y= np.sqrt(sum((y-y_mean)**2)/(len(y)-1))

cov_x_y=sum((x-x_mean)*(y-y_mean))/(len(x)-1)

print("Corelattion of x and y :",cov_x_y/(sigma_x*sigma_y))
```
