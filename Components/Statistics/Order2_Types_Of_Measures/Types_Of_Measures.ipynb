# Variable

__Variables__ are features what we are studying for our analysis.

__Variables__ should be Measureable,Countable,Categorized.

# Types of variables

1. __Quantitative__ : Data that is measured in __Numbers__ . But the Numbers only make sense to perform __Arithemetic Calculations__.

    a. here Variables are __Measureable,Countable but not Categorizeable__
    
    b. eg: height, weight, score, Age, Rank. In this examples variables are __Categorizeable__ when they are converted in to __ranges__, then they are considered as __Categorical__
    

2. __Qualitative | Categorical__ : Values that place things into __groups or category__
    
    a. **Arithemetic operations** dont make sence

    b. here Variables are __Categorizeable,Countable but not Measureable__
    
    c. eg: hair color, grade, spam|not spam, Age group, Rank Group.
    
    d. numbers can also consider as categorical dependent upon situation, like happiness index value start from 0 to 5, these are number but arithemetic operations dont make sence.
    

3. __interval__ : interval variables are also **Quantitative variables** can be ordered and have meaningful differences between values.

    a. **Arithemetic operations** make sence
    
    b. __NO natural zero point.__  natural zero point means zero represents absense of variable eg: 0 money,0 height,0 weight
    
    c. Also **Ratio** wont make sence

    d. For example, temperature in Celsius or Fahrenheit is an interval variable, because the difference between 10°C and 20°C is the same as the difference between 40°C and 50°C, but 0°C does not mean the absence of temperature.and 10°C is not the half temperature(cold or hot) of 20°C ratio dont work here.
    
    e. Interval measurement of data is **Quantitative data**
    
    
4. __Ratio__: A **ratio variable**, has all the properties of an **interval variable**, but also has a clear definition of 0.0.  And equal intervals between values.

    a. A true zero point means that the value of zero represents the absence of the variable, such as zero height, zero weight, or zero money.
    
    b. Equal intervals mean that the difference between any two values is the same, such as the difference between 10 cm and 20 cm is the same as the difference between 40 cm and 50 cm.
    
    c. Ratio data is the **highest level of measurement and can be used for all types of statistical analysis**, such as calculating the mean, median, mode, range, standard deviation, correlation, or regression. Some examples of ratio data are height, weight, age, speed, or income.
    
    d. Ratio measurement of data is **Quantitative data**
    
    e. **ratio is valid here**
    

# Types of Quantitative Variable

1. __Discrete__ : Refers to values that can only be measured in __Certain Numbers.__

    a. it's not about whole number,decimal number etc. its about the limitation of variable to have any numbers

    b. eg: Number of pets u have which cant be 1.5 or 2.5 it will be always integer.
    
    c. Even though it has limitation of variable to have any numbers **still it is a quantitave variable, when we considering a values as discrete make sure it can also do Arithemetic operation because it can also be a Categorical value.** 
    
    

2. __Continous__ : Refers to values that can take __any numerical values__ there is __no restriction__

    a. There is no limitation for having any Numerical value for a variable. 
    
    b. eg: weight.

# Types of Categorical Variable

1. __Ordinal__ : if there is a __Logical ordering__ to the values of a categorical variable its Ordinal.

    a. Logical ordering is not like alphabetical ordering, there must be a logical meaning 
    
    b. **Arithemetic operation** can't be done.
    
    b. eg: Grade of a class.A+ is greater than A. 
    
    

2. __Nominal__ : if there is __No Logical ordering__ to the values of a categorical variable its Nominal.
    
    a. eg: Hair colour.
    
    b. **Arithemetic operation** can't be done.

a. A line chart is, as one can imagine, a line or multiple lines showing how __single, or multiple variables develop over time.__

# Pie Chart

the total quantity it represents.

b. Its only have one variable and it will be __Categorical__

c. shows Percentage wise distribution of __Categorical__ data.

# Bar Graph

a. A chart that presents __categorical data__ with __rectangular bars__ with heights(for horizontal) or lengths(for vertical) proportional to the __values that they represent.__

b. The bars can be plotted vertically or horizontally. 

c. Can be of one variable or many variable.

d. For a normal Bar chart. __x axis: Categorical Variable VS y axis: Numerical Variable__ wise versa for vertical Bar chart

e. A bar chart is a graph that shows the frequency or relative frequency distribution of a __categorical variable__

f. In bar graphs, the order of the categories does not usually matter. However, bar graphs that have categories arranged in __decreasing order of frequency help prioritize categories for decision-making purposes(this chart is known as Pareto chart)__ in areas such as quality control, human resources, and marketing.

# Histogram

a. A series of bins showing us the frequency of observations of a given variable.

b. In each interval __end points are not included (Exclusive) and starting point is Inclusive__ and __each bin size should equal.__

c. It show frequency distribution of __single Numerical variable__.

d. x axis: Bins(converting Numerical to Categorical) VS y axis: Frequency of Variable as per bins

e. Variable is converted in to bins(categorical variable) and shows how many frequencies for each bins.

f. it shows spread of the data.

g. The histogram, a graph used to present quantitative data, is similar to the bar graph.

h. in histogram intervals will only created if value in continous or discrete with many values. If discrete variable with less values then we take variable itself in x bar instead of bins ( When a data set consists of a large 
number of different discrete data values or when a data set consists of continuous data, we must create classes by using intervals of numbers).
