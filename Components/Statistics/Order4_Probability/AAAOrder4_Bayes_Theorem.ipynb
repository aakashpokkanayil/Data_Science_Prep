# Bayes Theorem

1. Best way to find reverse Probability

2. It is basically Conditional Prob/ Total Prob

3. **Reverse Probability:**

    Normally we will try to find Probability of B (P(B)) by using Law of Total Probability.
    
    P(B)=P(A)*P(B|A)+P(A')*P(B|A')
    
    In **Reverse Probability:** Event B will be already happen. Here we are trying to find out P(A|B)
    
    P(A|B)=P(A and B)/P(B),  P(A and B)=P(B and A)
    
          = P(B and A) / P(B)  --- rewriting
          
          = (P(A) *P(B|A)) / P(B)
          
    P(A|B)= (P(A)*P(B|A) / P(B))
    
    For doing solutions easily:
    
    P(A|B)=Path to B via A / Total Prob of B

### Conditions

Since Bayes therem uses Law of total prob, Conditions for Law of Total Probs apply here too

1. Probability of every event >0

2. Every Events have to be Disjoint together.

3. Events should be Exhaustive.
