# What is Conditional Probability

ref: https://www.youtube.com/watch?v=_IgyaD7vOOA

1. Condition Probability measure the probability of one event occurring relative to another
occurring.

2. Conditional Probability states that P(A ∩ B|B) = P(A ∩ B)/P(B), means Probability of A and B where B is already happen.

3. Here sample space will be B, we are finding A ∩ B (or simply A) in sample space B

4. In short terms : **Very Imp Formulas**

    **P(A|B) = P(A ∩ B)/P(B)**

    **Multiplicative rule: P(A ∩ B) = P(B) * P(A|B)**

5. In summary P(A|B) and P(A ∩ B|B) have same meaning, Inside B ,A and A ∩ B are same.

6. Example: Probability of Rainbow when Raining.


# Multiplication Rule

### If A and B are **dependent events**, then  P(A ∩ B) = P(B) * P(A|B)

### If A and B are **Independent events**, then  P(A ∩ B) = P(A) * P(B)
