```python
import numpy as np
import pandas as pd

# Joining Cost 1$

JackpotGame=pd.DataFrame({
    'Conbinations':["winning nothing","all lemon","all cherries","$$cherry","$$$"],
    'Probabilities':[0.977,.008,.008,.006,.001],
    'Winning Price':[0,5,10,15,20],
    "Profit": [-1,4,9,14,19] # after considering Joining Cost
             })

JackpotGame
```

1. Linear Transformation method Is a shortcut way or easy way to identify new Expectation and Varience.

# Properties of Expectation and Varience

#### X and Y here two Indepedndent Random Variables (dont confuse with X and Y above). You can Take like Serial weight of box and serial weight of bowl

1. E(X + Y) = E(X) + E(Y)

2. E(X - Y) = E(X) - E(Y)

3. VAR(X + Y)= VAR(X) + VAR(Y)

4. VAR(X - Y)= VAR(X) + VAR(Y)

# Properties of Linear Transformation

1. E(aX + bY) = aE(X) + bE(Y)

2. E(aX - bY) = aE(X) - bE(Y)

3. VAR(aX + aY)= a^2VAR(X) + b^2VAR(Y)

4. VAR(aX - bY)= a^2VAR(X) + b^2VAR(Y)

### Now did'nt you got confused why VAR(X - Y)= VAR(X) + VAR(Y) not **VAR(X) - VAR(Y)** ?

lets take an example:

   X: is weight of serial box (serial box: its collection of weights)
    
   Y: is weight of serial bowl
    
   E(X) = 160 oz ,  E(Y) = 40 oz
    
   STD(X) = .80 oz , STD(Y) = .60 oz
    
   weight of X varries btw : 15 <= x >= 17 ---> (1)
    
   weight of Y Varries btw : 3 <= y >=5    ---> (2)
    
   **when we combine X and y in a way that range of the combination has to be maximum:**
    
    X+Y ==> 18 <= x+y >= 22 (this is the max range when adding X and Y)  ---> (3)
    
    X-Y ==> 10(15-5) <= x-y >= 14(17-3) (this is the max range when subs X and Y)   ---> (4)
    
    Now, X and Y has spread 2  ie 17-15=2 and 5-3=2
    
    and when we add X and Y, then spread is added up , ie spread of X + Y = Spread(X) + Spread(Y) = 4
    
    and when we subs X and Y, then spread is still added up , ie spread of X - Y = Spread(X) + Spread(Y) = 4
    
    thats why Var(X+Y) and Var(X-Y) are same at the end , VAR(X) + VAR(Y)

# Assignment Question.

It is asked Price Difference of both Restaurants so we an find differences of Expectations of both Restaurants

    E(X-Y) = E(X) - E(Y)
    
    VAR(X-Y) = VAR(X) + VAR(Y)

```python
RestaurantA=pd.DataFrame({
    'X':[20,30,40,45],
    'Prob':[.3,.4,.2,.1]
                     })


RestaurantB=pd.DataFrame({
    'Y':[10,15,18],
    'Prob':[.2,.6,.2]
                     })

```

```python
Expectation_A=sum(RestaurantA.X*RestaurantA.Prob)
print("Expectation of Restaurant A: ",Expectation_A)
```

```python
Expectation_B=sum(RestaurantB.Y*RestaurantB.Prob)
print("Expectation of Restaurant B: ",Expectation_B)
```

```python
# E(X-Y) = E(X) - E(Y)
print("differences of Expectations of both Restaurants: ",Expectation_A-Expectation_B)
```

```python
# VAR(X-Y) = VAR(X) + VAR(Y)

# E(X-Mue)^2=Σ((xi-E(X))^2 * Pi)

Varience_A=sum(((RestaurantA.X - Expectation_A)**2) * RestaurantA.Prob)
print("Varience of Restaurant A: ",Varience_A)

Varience_B=sum(((RestaurantB.Y - Expectation_B)**2) * RestaurantB.Prob)
print("Varience of Restaurant B: ",Varience_B)
```

```python
# VAR(X-Y) = VAR(X) + VAR(Y)
print("differences of Variences of both Restaurants: ",Varience_A-Varience_B)
```
