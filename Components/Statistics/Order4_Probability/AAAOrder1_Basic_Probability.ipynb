1. Probability is nothing but quantification of chances or possibilities

# Sample Space

1. Set of all possible outcomes. Denotes by **S**

2. Example: tossing a coin have a sample space of {H,T}

# Event

1. Desired or favourable outcome. Denotes by **E**

2. Example: getting Head when Tossing a coin is an event , here getting a coing is favourable outcome.

# Probability of an event

1. P(E)=n(E)/n(S), where 

   a. n(E) is No of Favarable outcomes
    
   b. n(S) is No of Sample Space
   
2. n(S)>=n(E) (always). So **Probability of an event P(E) will be between 0 to 1**

3. Examples of Events.

   a. Toss a coin and getting Head
      
      S= {H,T} , E={H} then P(H)= n(H)/n(S) = 1/2
      
   b. Drawing a Dice and getting 2 on dice
   
      S={1,2,3,4,5,6} , E={2} then P(2)=n(2)/n(S)= 1/6
      
   c. Drawing a Dice and getting value >=4
   
      S={1,2,3,4,5,6} , E={4,5,6}  then P(n>=4) = n(n>=4)/n(S)= 3/6==> 1/2
      
   d. Drawing a Dice and getting even value
   
      S={1,2,3,4,5,6} , E={2,4,6} then P(n=even number)= n(even number)/n(S) = 3/6 ==> 1/2
      
   e. Drawing a Dice and getting Prime value
   
      S={1,2,3,4,5,6} , E={2,3,5} then P(n=Prime number) = n(Prime number)/n(S)= 3/6 ==> 1/2

# Types of event

1. Complimentary Events

2. Mutually Exclusive Event

3. Non Mutually Exclusive Event

4. Exhaustive Event.

# Complimentary Events

1. if A and A' are complimentary events then P(A)+ P(A')=1

2. we can also write P(A')= 1- P(A)

# Exhaustive Event

1. If A and B are two events in a sample space, then if P(A U B)=1 then A and B are Exhaustive Events.

2. If A and B are Exhaustive Events then AUB is Sample Space itself.

3. Complementary events are exhaustive events.
