# Independent and Dependent Events

**Dependent Events:** Events A and B are said to be dependent if P(A | B) **(** P(A and B | B) **)** is different from P(A). It’s a way of saying that the probabilities of A and B are affected by each other.

when A and B are dependent P(A and B)= P(A | B) * P(B)

**Independent Events:** Independent events aren’t affected by each other. They don’t influence
each other’s probabilities in any way at all. If one event occurs, the probability of the other
occurring remains exactly the same.

So, P(A | B)  = P(A) itself.

In dependent event we can write P(A and B | B) is same as P(A | B) but Independent events both are different **try to visualise it with venn diagram.**

P(A and B)= P(A) * P(B)

# Law of Total Probability

**It Gives a way of Finding Total Probability of a particular Event (Total Prob of P(B)) based on Conditional Probability.**

**Conditions of Total Probability:**

1. Prob of every event must >0.

2. Events Must be disjoint.

3. adding all Events Must be Exhaustive




    

Law of Total probability is about Finding Total Probability of a particular Event

Suppose B is that event and A is the event happend before B

by using venn diagram we can say P(B)= P(B and A) + P(B and A')





P(B) = P(B and A) + P(B and A')

P(B) = P(A) * P(B | A) + P(A') * P(B | A')
